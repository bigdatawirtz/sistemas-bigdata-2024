```python
# 3.1 Show a graph of deaths in the three countries for the month of January, including titles, etc...
```

```python
datos_xaneiro = df_covid.sort_index().loc['2021-1-1':'2021-1-31'][['pais','mortes']]
df_spain = datos_xaneiro[datos_xaneiro.pais == 'Spain']
df_portugal = datos_xaneiro[datos_xaneiro.pais == 'Portugal']
df_france = datos_xaneiro[datos_xaneiro.pais == 'France']
plt.figure(figsize=(20,7))
plt.xlabel('data')
plt.ylabel('nÃºmero de mortes por dÃ­a')
sns.lineplot(data = df_spain.mortes,label='spain')
sns.lineplot(data = df_portugal.mortes,label='portugal')
sns.lineplot(data = df_france.mortes,label='france')
plt.title('EvoluciÃ³n das mortes en diferentes paÃ­ses')
```

```python
datos_xaneiro = df_covid.sort_index().loc['2021-1-1':'2021-1-31'][['pais','mortes']]
df_spain = datos_xaneiro[(datos_xaneiro.pais == 'Spain') & (datos_xaneiro.mortes != 0)]
df_portugal = datos_xaneiro[datos_xaneiro.pais == 'Portugal']
df_portugal = datos_xaneiro[(datos_xaneiro.pais == 'Portugal') & (datos_xaneiro.mortes != 0)]
df_france = datos_xaneiro[datos_xaneiro.pais == 'France']
df_france = datos_xaneiro[(datos_xaneiro.pais == 'France') & (datos_xaneiro.mortes != 0)]

plt.figure(figsize=(20,7))
plt.xlabel('data')
plt.ylabel('nÃºmero de mortes por dÃ­a')
sns.lineplot(data = df_spain.mortes,label='spain')
sns.lineplot(data = df_portugal.mortes,label='portugal')
sns.lineplot(data = df_france.mortes,label='france')
plt.title('EvoluciÃ³n das mortes en diferentes paÃ­ses')
```

```python
# 3.2 Try to change the color of the lines assign a new color to each country.
# In addition to changing the color increase the width of the lines
# https://seaborn.pydata.org/generated/seaborn.lineplot.html
```

```python
datos_xaneiro = df_covid.sort_index().loc['2021-1-1':'2021-1-31'][['pais','mortes']]
df_spain = datos_xaneiro[(datos_xaneiro.pais == 'Spain') & (datos_xaneiro.mortes != 0)]
df_portugal = datos_xaneiro[datos_xaneiro.pais == 'Portugal']
df_portugal = datos_xaneiro[(datos_xaneiro.pais == 'Portugal') & (datos_xaneiro.mortes != 0)]
df_france = datos_xaneiro[datos_xaneiro.pais == 'France']
df_france = datos_xaneiro[(datos_xaneiro.pais == 'France') & (datos_xaneiro.mortes != 0)]

plt.figure(figsize=(20,7))
plt.xlabel('data')
plt.ylabel('nÃºmero de mortes por dÃ­a')
sns.lineplot(data = df_spain.mortes, color='red',label='spain',linewidth=5)
sns.lineplot(data = df_portugal.mortes, color='purple',label='portugal', linewidth=3)
sns.lineplot(data = df_france.mortes, color='orange',label='france')
plt.title('EvoluciÃ³n das mortes en diferentes paÃ­ses')
```
